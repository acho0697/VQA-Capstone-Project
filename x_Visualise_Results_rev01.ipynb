```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
experiment_name = 'gqaExperiment'

data_location = './results/'+ experiment_name + '/'


# To load, you need to separate the specific experiment data from the results-gqaExperiment
data = pd.read_csv(data_location+'results-'+experiment_name+'.csv', header=1)

print(data)
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['epoch'], data['trainAcc'], label = 'training')
plt.plot(data['epoch'],  data['valAcc'], label = 'validation')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
```
