```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
experiment_name = 'gqaSG'

data_location = './results/'+ experiment_name + '/'


# To load, you need to separate the specific experiment data from the results-gqaExperiment
data = pd.read_csv(data_location+'results-'+experiment_name+'.csv', header=1)

print(data)
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['epoch'], data['evalTrainAcc'],  linestyle= ':', color='chocolate', label = 'training (eval)')
plt.plot(data['epoch'],  data['valAcc'], color='chocolate', label = 'validation')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
```

```python
experiment_name = 'gqaBaseline'

data_location = './results/'+ experiment_name + '/'


# To load, you need to separate the specific experiment data from the results-gqaExperiment
data2 = pd.read_csv(data_location+'results-'+experiment_name+'.csv', header=1)

print(data2)
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['epoch'], data['evalTrainAcc'], linestyle= ':', color='chocolate', label = 'SG training (eval)')
plt.plot(data['epoch'],  data['valAcc'], color='chocolate', label = 'SG validation')
plt.plot(data2['epoch'], data2['evalTrainAcc'], linestyle= ':', color='dimgray', label = 'Baseline training (eval)')
plt.plot(data2['epoch'],  data2['valAcc'], color='dimgray', label = 'Baseline validation')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
```
